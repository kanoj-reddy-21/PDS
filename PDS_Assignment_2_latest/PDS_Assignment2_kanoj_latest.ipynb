DATA CLEANING

a) Look for the missing values in all the columns and either impute them.

```python
import warnings

car_df = pd.DataFrame(data)

def parse_mileage(mileage):
    if pd.isnull(mileage):
        return np.nan
    return float(mileage.replace(' kmpl', '').replace(' km/kg', ''))

car_df['Mileage'] = car_df['Mileage'].apply(parse_mileage)


def parse_price(price):
    if pd.isnull(price):
        return np.nan
    price = price.replace(' Lakh', '').replace(' Cr', '')
    scale_factor = 1 if 'Lakh' in price else 100
    return float(price) * scale_factor


print(car_df.isnull().sum())


car_df['Mileage'].fillna(car_df['Mileage'].mean(), inplace=True)


car_df['Power'] = car_df['Power'].str.replace(' bhp', '').astype(float)
car_df['Power'].fillna(car_df['Power'].mean(), inplace=True)

car_df['Seats'].fillna(car_df['Seats'].mode()[0], inplace=True)


car_df['Engine'].fillna(car_df['Engine'].mode()[0], inplace=True)


car_df.drop(columns=['New_Price'], inplace=True)

print("After filling missing values:")
print(car_df.isnull().sum())

output_csv_file = '//content/drive/MyDrive/PDS/PDS_Assignment_2/cleaned_data.csv'

car_df.to_csv(output_csv_file, index=False)

print(f"Cleaned data has been saved to {output_csv_file}")
```

To standardize the dataset, we first converted the units to a uniform format for each relevant column.Missing values in the mileage and power columns were filled with their respective means, while the seats and engine columns used the most frequent (mode) values for imputation.Due to a large number of missing entries in the New_Price column, it was removed from the dataset.

b) Remove the units from some of the attributes and only keep the numerical value.

```python

car_df['Mileage'] = car_df['Mileage'].astype(str)
car_df['Engine'] = car_df['Engine'].astype(str)
car_df['Power'] = car_df['Power'].astype(str)

car_df['Mileage'] = car_df['Mileage'].str.extract('(\d+\.\d+|\d+)')[0].astype(float)
car_df['Engine'] = car_df['Engine'].str.replace(' CC', '').astype(float)
car_df['Power'] = car_df['Power'].str.extract('(\d+\.\d+|\d+)')[0].astype(float)



output_csv_file = '//content/drive/MyDrive/PDS/PDS_Assignment_2/units_data.csv'


car_df.to_csv(output_csv_file, index=False)

print(f"data has been saved to {output_csv_file}")
car_df.head()
```
