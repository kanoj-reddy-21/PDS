```python
import pandas as pd

#importing the raw data
data_frame = pd.read_csv('/content/drive/MyDrive/PDS/Q1/raw_data/raw_data.csv')
data_frame.head(10)

```

```python
#rename the file
data_frame.rename(columns = {'Height (Inches)':'height',
                       'Weight (Pounds)':'weight',
                       'Age':'age',
                        'Grip strength':'grip_strength',
                         'Frailty':'frailty'}, inplace = True)
data_frame.frailty=data_frame.frailty.map(dict(Y=1, N=0))
```

```python
#displaying conents to csv
data_frame.to_csv('/content/drive/MyDrive/PDS/Q1/clean_data/clean_data.csv')
print(data_frame)
```

Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

final=sns.histplot(data_frame['Height'], bins=10, kde=True)
plt.title('Height Distribution')
plt.show()
#Save the figure
fig = final.get_figure()
fig.savefig('/content/drive/MyDrive/PDS/Q1/result/height_distribution.png')
```

```python
# Create the bubble plot
plt.figure(figsize=(10, 6))
final = sns.scatterplot(data=data_frame, x='age', y='grip_strength',
                        size='grip_strength',
                        sizes=(50, 300),
                        hue='frailty',  # Color by frailty
                        palette='coolwarm',  # Choose a color palette
                        alpha=0.6)

# Adding titles and labels
plt.title('Bubble Plot of Grip Strength vs. Age')
plt.xlabel('Age')
plt.ylabel('Grip Strength')
plt.grid(True)

# Show the plot
plt.show()

# Save the figure
fig = final.get_figure()
fig.savefig('/content/drive/MyDrive/PDS/Q1/result/bubble_plot_grip_vs_age.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the box plot
plt.figure(figsize=(10, 6))
final=sns.boxplot(x='frailty', y='Weight', data=data_frame,palette={'0': 'red', '1': 'blue'})

# Adding titles and labels
plt.title('Box Plot of Weight by Frailty Status')
plt.xlabel('Frailty Status (1: Frail, 0: Not Frail)')
plt.ylabel('Weight (lb)')
plt.xticks([0, 1], ['Not Frail', 'Frail'])
plt.grid(True)

# Show the plot
plt.show()
fig = final.get_figure()
fig.savefig('/content/drive/MyDrive/PDS/Q1/result/box_plot_of_weight_by_frailty_status')
```
