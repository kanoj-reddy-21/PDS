```python
import pandas as pd

#importing the raw data
data_frame = pd.read_csv('/content/drive/MyDrive/PDS/Q2/raw_data/StudentsPerformance.csv')
data_frame.head(10)

```

```python
data_frame.rename(columns = {'race/ethnicity':'race_or_ethnicity',
                     'parental level of education':'parental_level_of_education',
    'test preparation course':'test_preparation_course',
                       'math score':'math_score',
                       'reading score':'reading_score',
                        'writing score':'writing_score'}, inplace = True)
```

```python
#displaying conents to csv
data_frame.to_csv('/content/drive/MyDrive/PDS/Q2/clean_data/StudentsPerformance_clean.csv')
print(data_frame)
```

Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

final=sns.histplot(data_frame['race_or_ethnicity'], bins=10)
plt.title('count of people in each groups')
plt.show()

#Save the figure
fig = final.get_figure()
fig.savefig('/content/drive/MyDrive/PDS/Q2/result/height_distribution.png')
```

```python
mean_scores = data_frame.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean()

diagram = mean_scores.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Test Preparation Status')
plt.ylabel('Average Score')
plt.xlabel('Test Preparation Status')
plt.xticks(rotation=0)  # Keeps the labels horizontal
plt.show()


# Save the figure
fig = diagram.get_figure()
fig.savefig('/content/drive/MyDrive/PDS/Q2/result/bar_graph_Ave_score_vs_prep_status.png')

```

```python
plt.figure(figsize=(14, 8))
diagram = sns.boxplot(data=data_frame, x='parental_level_of_education', y='math_score')
plt.title('Math Scores by Parental Level of Education')
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()



# Save the figure
fig = diagram.get_figure()
fig.savefig('/content/drive/MyDrive/PDS/Q2/result/box_plot_parent_level_edu_vs_math_score.png')

```

```python
import matplotlib.pyplot as plt

# Create a figure and axes object
fig, ax = plt.subplots()

# Plot the histogram on the axes object
ax.hist(data_frame['math_score'], bins=20, color='lightgreen', alpha=0.7, label='Math Scores')
ax.set_title('Distribution of Math Scores')
ax.set_xlabel('Score')
ax.set_ylabel('Frequency')
ax.legend()

# Display the plot
plt.show()

# Save the figure to a file
fig.savefig('/content/drive/MyDrive/PDS/Q2/result/distribution_of_math_score.png')
```

```python

# Create a figure and axes object
fig, ax = plt.subplots(figsize=(8, 8))

# Calculate lunch counts
lunch_counts = data_frame['lunch'].value_counts()

# Create a pie chart on the axes object
wedges, texts, autotexts = ax.pie(lunch_counts, labels=lunch_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
ax.set_title('Lunch Status Distribution')

# Display the plot
plt.show()

# Save the figure to a file
fig.savefig('/content/drive/MyDrive/PDS/Q2/result/piechart.png')
```

**Bar Graph of count of people in each groups:**

Analysis:It helps identify the number of people in each group giving analysis if data is properly distributed or not.

**Average Scores by Test Preparation Status:**

Analysis: Average score by test preparation status gives mean of the scores based on the status of test preparation. this tells the level of exam and amount of preparation required for each subject

**Box Plot of Scores by Parental Level of Education:**

Analysis: Reveals the distribution of scores across different levels of parental education. This can be used to understand trends in academic performance relative to parental education background, such as median scores and potential outliers.

**Histogram of Score Distributions:**

Analysis: Shows the frequency distribution of scores, making it clear how students are clustered around certain score ranges. This helps in assessing the overall performance and identifying common score ranges.

**Pie Chart of Lunch Status:**

Analysis: Provides a clear percentage breakdown of students' lunch status, highlighting the proportion of students who receive free/reduced lunch versus a standard lunch, which can be a proxy for socio-economic status.

**Scatter Plot of Math vs. Writing Scores**

Analysis: Helps in examining the relationship between performance in math and writing, indicating correlation and dispersion of scores. This can reveal patterns such as whether high achievers in math are also high achievers in writing.
